### Question 5

```r
getwd()

pat_names <- read.csv("Homework_3/patient_names.csv")
pat_props <- read.csv("Homework_3/patient_properties.csv")

pat_names1 <- pat_names %>%
    mutate(
        BIRTHDATE = as.Date(BIRTHDATE, format = "%m/%d/%y"),
        DEATHDATE = as.Date(DEATHDATE, format = "%m/%d/%y")
    )

head(pat_names1, 10)
head(pat_props, 10)
```

### Question 6

```r
library(dplyr)
library(tidyr)

pat_props_wide <- pat_props %>%
    pivot_wider(
        names_from = property,
        values_from = value
    )

head(pat_props_wide,10)
```

### Question 7

```r
pat_join <- pat_names1 %>%
    left_join(
        pat_props_wide,
        by = "ID"
    )

head(pat_join,10)
```

### Question 8

```r
#Each first and last name has 3 digits stuck on the end. They need to be trimmed

pat_join$FIRST <- gsub("[0-9]+$", "", pat_join$FIRST)
pat_join$LAST <- gsub("[0-9]+$", "", pat_join$LAST)

head(pat_join,10)
```

### Question 9

```r
p_cat_cols <- setdiff(names(pat_join), c("ID", "FIRST", "LAST"))
for (var in p_cat_cols) {
    if(is.character(pat_join[[var]]) || is.factor(pat_join[[var]]) || inherits(pat_join[[var]], "Date")) {
        counts <- pat_join %>%
        count(.data[[var]]) %>%
        arrange(desc(n))
        cat("\n###", var, "\n")
            mdpre(counts)
        }
    }
```

### Question 10

Variables to clean:  
MARITAL: 1 "Fine", 1 "male", 1 NA. Trim the illogical values.  
RACE: 1 "asiann". Combine into "asian"  
ETHNICITY: 1 "nonhispani", 1 "hispani". Combine into "nonhispanic" and "hispanic" respectively.  
GENDER: 1 Female, 1 female, 1 Male. Combine into appropriate F or M category  

```r
pat_join <- pat_join %>%
    filter(!(MARITAL %in% c("Fine","male")))
pat_join$RACE[pat_join$RACE %in% c("asiann")] <- "asian"
pat_join$ETHNICITY[pat_join$ETHNICITY %in% c("nonhispani")] <- "nonhispanic"
pat_join$ETHNICITY[pat_join$ETHNICITY %in% c("hispani")] <- "hispanic" 
pat_join$GENDER[pat_join$GENDER %in% c("female","Female")] <- "F"
pat_join$GENDER[pat_join$GENDER %in% c("Male")] <- "M"
```

```r
pat_join <- pat_join %>%
    mutate(
            BIRTHDATE = if_else(year(BIRTHDATE) >= 2000 & year(BIRTHDATE) <= 2099,
                    BIRTHDATE %m-% years(100),
                    BIRTHDATE)
        )

head(pat_join,10)
```

### Question 11

```r
pat_join <- pat_join %>%
    mutate(
        age = ifelse(
            is.na(DEATHDATE),
            floor(as.numeric(difftime(Sys.Date(), BIRTHDATE, units = "days")) / 365.25),
            floor(as.numeric(difftime(DEATHDATE,BIRTHDATE, units = "days")) / 365.25)
        ))
```

```r
ggplot(pat_join, aes(x = age, fill = GENDER)) +
    geom_histogram(position = "dodge", bins = 30) +
    labs(
            title = "Histogram of Patient Ages by Gender",
            x = "Age (years)",
            y = "Count"
        )
```

```r
ggplot(pat_join, aes(x = MARITAL, y = BIRTHDATE, color = MARITAL)) +
    geom_jitter(alpha = 0.7, width = 0.35, height = 0.2) +
    labs(
        title = "Birthdates by Marital Status",
        x = "Marital Status",
        y = "Birthdate"
    )
```
